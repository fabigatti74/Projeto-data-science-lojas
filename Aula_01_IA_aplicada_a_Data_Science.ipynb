```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Carregar base de clientes
url_clientes = "https://raw.githubusercontent.com/alura-cursos/ia-datascience-dataviz/main/dados/zoop_clientes.csv"
df_clientes = pd.read_csv(url_clientes)

# Carregar base de vendas
url_vendas = "https://raw.githubusercontent.com/alura-cursos/ia-datascience-dataviz/main/dados/zoop_vendas.csv"
df_vendas = pd.read_csv(url_vendas)

# Visualizar as primeiras linhas
print("Clientes:")
print(df_clientes.head())

print("\nVendas:")
print(df_vendas.head())
```

```python
# Converter a coluna 'data' para datetime
df_vendas['data'] = pd.to_datetime(df_vendas['data'])

# Filtrar vendas de 2023
vendas_2023 = df_vendas[df_vendas['data'].dt.year == 2023]
```

```python
pagamento_mais_usado = vendas_2023['metodo_pagamento'].value_counts()

print("Métodos de pagamento mais utilizados em 2023:")
print(pagamento_mais_usado)

# Gráfico
pagamento_mais_usado.plot(kind='bar')
plt.title('Métodos de pagamento mais utilizados em 2023')
plt.xlabel('Método de Pagamento')
plt.ylabel('Quantidade')
plt.show()
```

```python
pagamento_mais_usado = vendas_2023['metodo_pagamento'].value_counts()

print("Métodos de pagamento mais utilizados em 2023:")
print(pagamento_mais_usado)

# Gráfico
pagamento_mais_usado.plot(kind='bar')
plt.title('Métodos de pagamento mais utilizados em 2023')
plt.xlabel('Método de Pagamento')
plt.ylabel('Quantidade')
plt.show()
```

```python
# Criar coluna 'valor_total'
vendas_2023['valor_total'] = vendas_2023['preco_unitario'] * vendas_2023['quantidade'] + vendas_2023['frete']

# Faturamento por categoria
faturamento_categoria = vendas_2023.groupby('categoria')['valor_total'].sum().sort_values(ascending=False)

print("Faturamento por categoria em 2023:")
print(faturamento_categoria)

# Gráfico
faturamento_categoria.plot(kind='bar')
plt.title('Faturamento por categoria em 2023')
plt.xlabel('Categoria')
plt.ylabel('Faturamento')
plt.show()
```

```python
# Criar coluna 'mes'
vendas_2023['mes'] = vendas_2023['data'].dt.month

# Faturamento por mês
faturamento_mes = vendas_2023.groupby('mes')['valor_total'].sum()

print("Faturamento por mês em 2023:")
print(faturamento_mes)

# Gráfico
faturamento_mes.plot(kind='bar')
plt.title('Faturamento mensal em 2023')
plt.xlabel('Mês')
plt.ylabel('Faturamento')
plt.show()
```

```python
# Criar coluna 'trimestre'
vendas_2023['trimestre'] = vendas_2023['data'].dt.to_period('Q')

# Pivot table
tabela_trimestre_pagamento = vendas_2023.pivot_table(
    index='trimestre',
    columns='metodo_pagamento',
    values='valor_total',
    aggfunc='sum'
)

print("Faturamento por trimestre e método de pagamento:")
print(tabela_trimestre_pagamento)

# Gráfico
tabela_trimestre_pagamento.plot(kind='bar', stacked=True)
plt.title('Faturamento por trimestre e método de pagamento (2023)')
plt.xlabel('Trimestre')
plt.ylabel('Faturamento')
plt.legend(title='Método de pagamento')
plt.show()
```

```python
# ------------------------------------------------------------------------------------
# Análise Exploratória — Base Zoop — Fabiana Gatti
# ------------------------------------------------------------------------------------

# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------------------------------
# 1️⃣ Carregamento das bases de dados
# ------------------------------------------------------------------------------------

# Base de clientes
url_clientes = "https://raw.githubusercontent.com/alura-cursos/ia-datascience-dataviz/main/dados/zoop_clientes.csv"
df_clientes = pd.read_csv(url_clientes)

# Base de vendas
url_vendas = "https://raw.githubusercontent.com/alura-cursos/ia-datascience-dataviz/main/dados/zoop_vendas.csv"
df_vendas = pd.read_csv(url_vendas)

# Visualização inicial
print("Clientes:")
print(df_clientes.head())

print("\nVendas:")
print(df_vendas.head())

# ------------------------------------------------------------------------------------
# 2️⃣ Preparação da base de vendas
# ------------------------------------------------------------------------------------

# Converter coluna 'data' para datetime
df_vendas['data'] = pd.to_datetime(df_vendas['data'])

# Filtrar vendas de 2023
vendas_2023 = df_vendas[df_vendas['data'].dt.year == 2023]

# Criar coluna 'valor_total'
vendas_2023['valor_total'] = vendas_2023['preco_unitario'] * vendas_2023['quantidade'] + vendas_2023['frete']

# ------------------------------------------------------------------------------------
# 3️⃣ Pergunta 1 — Métodos de pagamento mais utilizados em 2023
# ------------------------------------------------------------------------------------

pagamento_mais_usado = vendas_2023['metodo_pagamento'].value_counts()

# Gráfico
pagamento_mais_usado.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Métodos de pagamento mais utilizados em 2023')
plt.xlabel('Método de pagamento')
plt.ylabel('Quantidade de transações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 4️⃣ Pergunta 2 — Faturamento por categoria em 2023
# ------------------------------------------------------------------------------------

faturamento_categoria = vendas_2023.groupby('categoria')['valor_total'].sum().sort_values(ascending=False)

# Gráfico
faturamento_categoria.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Faturamento por categoria em 2023')
plt.xlabel('Categoria')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 5️⃣ Pergunta 3 — Vendas ao longo dos meses em 2023
# ------------------------------------------------------------------------------------

# Criar coluna 'mes'
vendas_2023['mes'] = vendas_2023['data'].dt.month

# Faturamento por mês
faturamento_mes = vendas_2023.groupby('mes')['valor_total'].sum()

# Gráfico
faturamento_mes.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Faturamento mensal em 2023')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 6️⃣ Pergunta 4 — Vendas por trimestre por método de pagamento
# ------------------------------------------------------------------------------------

# Criar coluna 'trimestre'
vendas_2023['trimestre'] = vendas_2023['data'].dt.to_period('Q')

# Pivot table
tabela_trimestre_pagamento = vendas_2023.pivot_table(
    index='trimestre',
    columns='metodo_pagamento',
    values='valor_total',
    aggfunc='sum'
)

# Gráfico
tabela_trimestre_pagamento.plot(kind='bar', stacked=True, colormap='tab20')
plt.title('Faturamento por trimestre e método de pagamento (2023)')
plt.xlabel('Trimestre')
plt.ylabel('Faturamento (R$)')
plt.legend(title='Método de pagamento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 7️⃣ Pergunta 5 — Proporção de clientes no programa de cashback
# ------------------------------------------------------------------------------------

# Verificar valores únicos
print("Valores únicos da coluna 'cashback':")
print(df_clientes['cashback'].unique())

# Calcular proporção
proporcao_cashback = df_clientes['cashback'].value_counts(normalize=True) * 100

# Gráfico de pizza
proporcao_cashback.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6), colors=['gold', 'lightgray'])
plt.title('Proporção de clientes no programa de cashback')
plt.ylabel('')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 8️⃣ Pergunta 6 — Distribuição de idade dos clientes
# ------------------------------------------------------------------------------------

df_clientes['idade'].plot(kind='hist', bins=10, edgecolor='black', color='lightblue')
plt.title('Distribuição de idade dos clientes')
plt.xlabel('Idade')
plt.ylabel('Quantidade de clientes')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 9️⃣ Pergunta 7 — Distribuição por sexo biológico
# ------------------------------------------------------------------------------------

sexo_dist = df_clientes['sexo_biologico'].value_counts()

# Gráfico
sexo_dist.plot(kind='bar', color='violet', edgecolor='black')
plt.title('Distribuição por sexo biológico')
plt.xlabel('Sexo biológico')
plt.ylabel('Quantidade de clientes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
