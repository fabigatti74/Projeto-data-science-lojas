### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 5. Frete Médio por Loja


```python
frete_medio_por_loja = dados.groupby('loja')['Frete'].mean()
print(frete_medio_por_loja)
```

```python
import matplotlib.pyplot as plt

# Criar o gráfico de barras
ax = frete_medio_por_loja.plot(kind='bar', figsize=(8,5))

# Adicionar título e rótulos
plt.title('Frete Médio por Loja')
plt.xlabel('Loja')
plt.ylabel('Valor Médio do Frete (R$)')
plt.xticks(rotation=0)
plt.tight_layout()

# Adicionar os valores em cima das barras
for barra in ax.patches:
    ax.text(
        barra.get_x() + barra.get_width() / 2,
        barra.get_height() + 0.5,
        f'R$ {barra.get_height():.2f}',
        ha='center',
        va='bottom',
        fontsize=9
    )

# Mostrar o gráfico
plt.show()
```

```python
# Faturamento total por loja
faturamento_por_loja = dados.groupby('loja')['Preço'].sum()

# Avaliação média por loja
avaliacao_por_loja = dados.groupby('loja')['Avaliação da compra'].mean()

# Frete médio por loja
frete_por_loja = dados.groupby('loja')['Frete'].mean()

# Volume total de vendas por loja (número de vendas)
volume_vendas_por_loja = dados['loja'].value_counts()

# Criar um DataFrame consolidado
comparativo_lojas = pd.DataFrame({
    'Faturamento': faturamento_por_loja,
    'Média Avaliação': avaliacao_por_loja,
    'Frete Médio': frete_por_loja,
    'Volume de Vendas': volume_vendas_por_loja
})

# Mostrar o comparativo
print('Comparativo entre as lojas:')
print(comparativo_lojas)

# Definir critério principal para decisão
# => Loja que tem MENOR faturamento
loja_para_fechar = faturamento_por_loja.idxmin()

print('\n>>> RECOMENDAÇÃO FINAL <<<')
print(f'Recomendamos o fechamento da {loja_para_fechar}, pois apresenta o pior desempenho de faturamento, apesar de ter frete competitivo e avaliação razoável.')
```

Avaliar o desempenho das quatro lojas da empresa **Alura Store**, com base em dados de vendas, avaliação de clientes e custo médio de frete, para recomendar qual unidade deve ser fechada visando a otimização dos recursos e melhores resultados.

---

## 2. Metodologia

Foram analisados os seguintes indicadores para cada loja:

- **Faturamento Total** (somatório das vendas)
- **Média das Avaliações dos Clientes**
- **Valor Médio do Frete**
- **Volume Total de Vendas**
- **Análise das Categorias de Produtos Vendidos**

As análises foram realizadas utilizando as bibliotecas **Pandas** e **Matplotlib** no Google Colab para tratamento e visualização dos dados.

---

## 3. Análises e Resultados

| Loja 4 | 1.384.497,58 |

- A **Loja 4** apresentou o **menor faturamento** entre as unidades analisadas.

- A **Loja 4** teve menor destaque nas principais categorias de alto volume de vendas.

## 4. Conclusão

Após análise detalhada dos indicadores, conclui-se que a **Loja 4** apresenta:

- O **menor faturamento**,
- Avaliação média apenas razoável,
